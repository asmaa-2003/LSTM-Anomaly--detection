```python
# =========================================================
# 🚀 SETUP
# =========================================================
import os
os.environ["TF_ENABLE_ONEDNN_OPTS"] = "0"

import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import precision_score, recall_score, f1_score, precision_recall_curve
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

print("GPU:", tf.config.list_physical_devices('GPU'))
```

```python
# =========================================================
# ⬇️ DOWNLOAD DATA
# =========================================================
DATA_PATH = "MetroPT2.csv"

!wget -q https://zenodo.org/records/7766691/files/MetroPT2.csv?download=1 -O {DATA_PATH}
print("✅ Downloaded")
```

```python
# =========================================================
# ⚙️ SETTINGS
# =========================================================
CHUNK_SIZE = 50000
TIMESTEPS = 30
BATCH_SIZE = 128
EPOCHS = 1

scaler = MinMaxScaler()

def create_sequences(data, timesteps):
    seqs = []
    for i in range(len(data) - timesteps):
        seqs.append(data[i:i+timesteps])
    return np.array(seqs)
```

```python
# =========================================================
# 🧠 FIXED LSTM AUTOENCODER (IMPORTANT 🔥)
# =========================================================

from tensorflow.keras.layers import RepeatVector, TimeDistributed

model = Sequential([
    LSTM(64, return_sequences=False, input_shape=(TIMESTEPS, 20)),
    Dropout(0.2),

    RepeatVector(TIMESTEPS),

    LSTM(64, return_sequences=True),
    Dropout(0.2),

    TimeDistributed(Dense(20))
])

model.compile(optimizer="adam", loss="mse")
model.summary()
```

```python
# =========================================================
# 🚀 TRAINING WITH CHUNKS
# =========================================================

print("🚀 Training started...")

for i, chunk in enumerate(pd.read_csv(DATA_PATH, chunksize=CHUNK_SIZE)):

    if chunk.columns[0].lower() in ["unnamed: 0", "index"]:
        chunk = chunk.drop(chunk.columns[0], axis=1)

    chunk["timestamp"] = pd.to_datetime(chunk["timestamp"])
    chunk = chunk.sort_values("timestamp")

    # anomalies
    chunk["is_anomaly"] = 0
    failure_windows = [
        ("2022-06-01", "2022-06-02"),
        ("2022-07-10", "2022-07-11")
    ]

    for start, end in failure_windows:
        mask = (chunk["timestamp"] >= start) & (chunk["timestamp"] <= end)
        chunk.loc[mask, "is_anomaly"] = 1

    feature_cols = chunk.select_dtypes(include=[np.number]).columns.tolist()
    if "is_anomaly" in feature_cols:
        feature_cols.remove("is_anomaly")

    X_chunk = chunk[feature_cols].values.astype(np.float32)

    # تدريب فقط على normal
    X_train = X_chunk[chunk["is_anomaly"] == 0]

    if len(X_train) < TIMESTEPS:
        continue

    scaler.fit(X_train)
    X_scaled = scaler.transform(X_train)

    X_seq = create_sequences(X_scaled, TIMESTEPS)

    if len(X_seq) == 0:
        continue

    model.fit(
        X_seq, X_seq,
        epochs=EPOCHS,
        batch_size=BATCH_SIZE,
        verbose=0
    )

    print(f"✅ Chunk {i+1} trained")

    # 🔥 تحرير RAM
    del chunk, X_chunk, X_train, X_scaled, X_seq

print("🎯 Training finished!")
```

```python
# =========================================================
# 📊 EVALUATION WITH CHUNKS
# =========================================================

errors = []
y_true_all = []

for i, chunk in enumerate(pd.read_csv(DATA_PATH, chunksize=CHUNK_SIZE)):

    chunk["timestamp"] = pd.to_datetime(chunk["timestamp"])
    chunk = chunk.sort_values("timestamp")

    chunk["is_anomaly"] = 0
    failure_windows = [
        ("2022-06-01", "2022-06-02"),
        ("2022-07-10", "2022-07-11")
    ]

    for start, end in failure_windows:
        mask = (chunk["timestamp"] >= start) & (chunk["timestamp"] <= end)
        chunk.loc[mask, "is_anomaly"] = 1

    feature_cols = chunk.select_dtypes(include=[np.number]).columns.tolist()
    if "is_anomaly" in feature_cols:
        feature_cols.remove("is_anomaly")

    X_chunk = chunk[feature_cols].values.astype(np.float32)
    X_scaled = scaler.transform(X_chunk)

    X_seq = create_sequences(X_scaled, TIMESTEPS)
    y_true = chunk["is_anomaly"].values[TIMESTEPS:]

    if len(X_seq) == 0:
        continue

    pred = model.predict(X_seq, verbose=0)
    error = np.mean(np.abs(X_seq - pred), axis=(1,2))

    errors.extend(error)
    y_true_all.extend(y_true)

    del chunk, X_chunk, X_scaled, X_seq, pred

errors = np.array(errors)
y_true_all = np.array(y_true_all)

print("✅ Evaluation data ready")
```

```python
# =========================================================
# 🔥 SMART THRESHOLD
# =========================================================

precision_vals, recall_vals, thresholds = precision_recall_curve(y_true_all, errors)

f1_scores = 2 * (precision_vals * recall_vals) / (precision_vals + recall_vals + 1e-8)
best_threshold = thresholds[np.argmax(f1_scores)]

print("Initial threshold:", best_threshold)
```

```python
# =========================================================
# 🚀 ULTRA THRESHOLD BOOST
# =========================================================

best_f1 = 0
best_threshold_final = 0
best_precision = 0
best_recall = 0

for m in np.linspace(1.0, 4.0, 50):

    thr = best_threshold * m
    y_pred = (errors > thr).astype(int)

    p = precision_score(y_true_all, y_pred, zero_division=0)
    r = recall_score(y_true_all, y_pred, zero_division=0)
    f1 = f1_score(y_true_all, y_pred, zero_division=0)

    if f1 > best_f1:
        best_f1 = f1
        best_threshold_final = thr
        best_precision = p
        best_recall = r

print("\n🔥 FINAL RESULTS:")
print(f"Threshold: {best_threshold_final:.6f}")
print(f"Precision: {best_precision:.4f}")
print(f"Recall:    {best_recall:.4f}")
print(f"F1 Score:  {best_f1:.4f}")
```
