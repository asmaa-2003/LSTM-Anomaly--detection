```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import precision_score, recall_score, f1_score
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, RepeatVector, TimeDistributed, Dense, Dropout

print("✅ Libraries imported successfully")
```

```python
DATA_PATH = "MetroPT2.csv"
CHUNK_SIZE = 50000
TIMESTEPS = 10
BATCH_SIZE = 32
scaler = StandardScaler()

print(f"✅ Settings done: TIMESTEPS={TIMESTEPS}, BATCH_SIZE={BATCH_SIZE}")
```

```python
first_chunk = next(pd.read_csv(DATA_PATH, chunksize=CHUNK_SIZE))
if first_chunk.columns[0].lower() in ["unnamed: 0", "index"]:
    first_chunk = first_chunk.drop(first_chunk.columns[0], axis=1)

feature_cols = first_chunk.select_dtypes(include=[np.number]).columns.tolist()
if "is_anomaly" in feature_cols:
    feature_cols.remove("is_anomaly")

n_features = len(feature_cols)
print("📊 Number of features:", n_features)
```

```python
def create_sequences(data, timesteps=TIMESTEPS):
    Xs = []
    for i in range(len(data) - timesteps):
        Xs.append(data[i:i+timesteps])
    return np.array(Xs)

print("✅ Sequence function ready")
```

```python
inputs = Input(shape=(TIMESTEPS, n_features))
x = LSTM(64, activation="tanh", return_sequences=False)(inputs)
x = Dropout(0.3)(x)
x = RepeatVector(TIMESTEPS)(x)
x = LSTM(64, activation="tanh", return_sequences=True)(x)
outputs = TimeDistributed(Dense(n_features))(x)

model = Model(inputs, outputs)
model.compile(optimizer="adam", loss="mse")
model.summary()
```

```python
print("\n🚀 Training using chunks...")

for i, chunk in enumerate(pd.read_csv(DATA_PATH, chunksize=CHUNK_SIZE)):
    if chunk.columns[0].lower() in ["unnamed: 0", "index"]:
        chunk = chunk.drop(chunk.columns[0], axis=1)

    chunk["timestamp"] = pd.to_datetime(chunk["timestamp"])
    chunk = chunk.sort_values("timestamp")

    chunk["is_anomaly"] = 0
    failure_windows = [
        ("2022-06-01", "2022-06-02"),
        ("2022-07-10", "2022-07-11")
    ]
    for start, end in failure_windows:
        mask = (chunk["timestamp"] >= start) & (chunk["timestamp"] <= end)
        chunk.loc[mask, "is_anomaly"] = 1

    feature_cols = chunk.select_dtypes(include=[np.number]).columns.tolist()
    if "is_anomaly" in feature_cols:
        feature_cols.remove("is_anomaly")
    X_chunk = chunk[feature_cols].values.astype(np.float32)

    train_mask = (chunk["timestamp"] < "2022-06-01") & (chunk["is_anomaly"] == 0)
    X_train_chunk = X_chunk[train_mask]

    if len(X_train_chunk) < TIMESTEPS:
        continue

    scaler.fit(X_train_chunk)
    X_train_scaled = scaler.transform(X_train_chunk)
    X_seq = create_sequences(X_train_scaled, TIMESTEPS)
    if len(X_seq) == 0:
        continue

    model.fit(
        X_seq, X_seq,
        epochs=1,
        batch_size=BATCH_SIZE,
        verbose=0
    )
    print(f"✅ Trained on chunk {i+1}: {len(X_seq)} sequences")

    # حفظ النموذج كل chunk لتجنب فقدانه
    model.save(f"/content/lstm_autoencoder_model_chunk{i+1}.keras")

print("\n🎯 Training completed without RAM crash!")
```

```python
model.save("/content/lstm_autoencoder_model_final.keras")
print("💾 Model saved!")
```

```python
# =========================
# تقييم باستخدام عدة chunks (تحسين Evaluation)
# =========================

chunks = []
reader = pd.read_csv(DATA_PATH, chunksize=CHUNK_SIZE)

# خذ أول chunks التي تشمل تاريخ 2022-06-01 أو بعده
for chunk in reader:
    chunk["timestamp"] = pd.to_datetime(chunk["timestamp"])
    # احتفظ فقط بالchunks التي تحتوي على بيانات بعد 2022-06-01
    if chunk["timestamp"].max() >= pd.Timestamp("2022-06-01"):
        chunks.append(chunk)
    if len(chunks) >= 3:  # خذ 3 chunks كحد أدنى
        break

# دمج جميع chunks معًا
sample_chunk = pd.concat(chunks)

# تنظيف
if sample_chunk.columns[0].lower() in ["unnamed: 0", "index"]:
    sample_chunk = sample_chunk.drop(sample_chunk.columns[0], axis=1)

sample_chunk = sample_chunk.sort_values("timestamp")

# إضافة anomalies فقط للتقييم
sample_chunk["is_anomaly"] = 0
failure_windows = [
    ("2022-06-01", "2022-06-02"),
    ("2022-07-10", "2022-07-11")
]
for start, end in failure_windows:
    mask = (sample_chunk["timestamp"] >= start) & (sample_chunk["timestamp"] <= end)
    sample_chunk.loc[mask, "is_anomaly"] = 1

# y_true بعد الانتهاء من وضع anomalies
y_true = sample_chunk["is_anomaly"].values[TIMESTEPS:]

# اختيار الميزات
feature_cols = sample_chunk.select_dtypes(include=[np.number]).columns.tolist()
if "is_anomaly" in feature_cols:
    feature_cols.remove("is_anomaly")

# scale و create sequences
X_sample = scaler.transform(sample_chunk[feature_cols].values.astype(np.float32))
X_seq_sample = create_sequences(X_sample, TIMESTEPS)

# التنبؤ و حساب reconstruction error
pred = model.predict(X_seq_sample, verbose=0)
error = np.mean((X_seq_sample - pred)**2, axis=(1,2))

# =========================
# اختيار threshold تلقائي بناءً على error
# =========================
# نرسم الـ error أولاً لرؤية قيمه
import matplotlib.pyplot as plt

plt.figure(figsize=(12,4))
plt.plot(error)
plt.title("Reconstruction Error")
plt.grid()
plt.show()

# threshold: نختاره حسب شكل error لتجنب أن تكون كل القيم صفر
threshold = np.percentile(error, 70)  # يمكنك تعديل 80% إلى 70% أو 75% إذا لم يظهر anomalies
y_pred = (error > threshold).astype(int)

# عرض عدد anomalies للتحقق
print("عدد الـ anomalies الحقيقية:", np.sum(y_true))
print("عدد الـ anomalies المكتشفة:", np.sum(y_pred))

# المقاييس
from sklearn.metrics import precision_score, recall_score, f1_score

precision = precision_score(y_true, y_pred, zero_division=0)
recall = recall_score(y_true, y_pred, zero_division=0)
f1 = f1_score(y_true, y_pred, zero_division=0)

print("\n📊 Evaluation Metrics:")
print(f"Precision: {precision:.4f}")
print(f"Recall:    {recall:.4f}")
print(f"F1 Score:  {f1:.4f}")
```

```python
print("\n⚠️ Early Warning Alerts:")

for i in range(2, len(error)):
    if error[i] > threshold and error[i] > error[i-1] > error[i-2]:
        print(f"⚠️ Strong failure signal at index {i}")

plt.figure(figsize=(12,5))
plt.plot(error, label="Reconstruction Error")
plt.axhline(threshold, color='r', linestyle='--', label="Threshold")
plt.title("Reconstruction Error with Anomalies")
plt.xlabel("Sequence Index")
plt.ylabel("Error")
plt.legend()
plt.grid()
plt.show()

print("\n✅ Pipeline completed successfully!")
```
